```python
comp1=[]
for i in range(len(comp)):
    comp1.append(comp['permalink'][i].split('/')[2])
```

```python
com1=[]
for i in range(len(rounds2)):
    com1.append(rounds2['company_permalink'][i].split('/')[2])
rounds2['company name']=com1
```

```python
comp['company name'] = comp["company name"].str.lower().str.strip()
rounds2['company name'] = rounds2["company name"].str.lower().str.strip()
print((comp['company name']).nunique())
print((rounds2['company name']).nunique())
```

```python
#merge company and rounds file using common column 'company name'
merge_data=pd.merge(comp,rounds2,on='company name',how='inner')
```
